```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, color='red')
```

```python
plt.plot(x,y, color='red', linewidth=5)
```

```python
plt.plot(x,y, color='red', linewidth=5, linestyle='dashed')
```

```python
plt.plot(x,y, color='red', linewidth=5, linestyle='dashdot')
```

```python
plt.plot(x,y, color='red', linewidth=5, linestyle='dashdot')
plt.xlabel('Day')
plt.ylabel('Temp')
plt.title("Weather")
```
