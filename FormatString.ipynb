```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, color='green', marker='+')
```

```python
plt.plot(x,y, 'g+') #g for green and + is marker
```

```python
plt.plot(x,y, 'g--') #order doesn't matter
```

```python
plt.plot(x,y, 'rD') #r means red and D is Diamond
```

```python
plt.plot(x,y, color='red', marker='D', linestyle='') 
```

```python
plt.plot(x,y, color='#000290', marker='D', linestyle='') 
```

```python
plt.plot(x,y, color='#000290', marker='D', markersize=20, linestyle='') 
```

```python
plt.plot(x,y, alpha=0) 
```

```python
plt.plot(x,y, alpha=0.5) 
```

```python
plt.plot(x,y, alpha=1) 
```
