```python
import matplotlib.pyplot as plt
import numpy as np
x=['python','c','c++','java']
y=[85,70,60,82]
z=[20,30,40,50]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("Languages", fontsize=15)
plt.ylabel("Number")
plt.title("Bar Plot")

plt.bar(p,y, width, color="r", label="BarPlotLevel")
plt.bar(p1,z, width, color="g", label="BarPlotLevel1")

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=['python','c','c++','java']
y=[85,70,60,82]
z=[20,30,40,50]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("Languages", fontsize=15)
plt.ylabel("Number")
plt.title("Bar Plot")

plt.bar(p,y, width, color="r", label="BarPlotLevel")
plt.bar(p1,z, width, color="g", label="BarPlotLevel1")

plt.xticks(p+width/2,x)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=['python','c','c++','java']
y=[85,70,60,82]
z=[20,30,40,50]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("Languages", fontsize=15)
plt.ylabel("Number")
plt.title("Bar Plot")

plt.bar(p,y, width, color="r", label="BarPlotLevel")
plt.bar(p1,z, width, color="g", label="BarPlotLevel1")

plt.xticks(p+width/2,x, rotation=20)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=['python','c','c++','java']
y=[85,70,60,82]
z=[20,30,40,50]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("Languages", fontsize=15)
plt.ylabel("Number")
plt.title("Bar Plot")

plt.barh(p,y, width, color="r", label="BarPlotLevel")
plt.barh(p1,z, width, color="g", label="BarPlotLevel1")

plt.legend()
plt.show()
```
