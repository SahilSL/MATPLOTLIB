```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,70]
plt.boxplot(x, widths=0.5, labels=['Python'])
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,70]
plt.boxplot(x, widths=0.1, labels=['Python'], patch_artist=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,70,120]
plt.boxplot(x, widths=0.1, labels=['Python'], patch_artist=True, showmeans=True)
plt.show()
```
