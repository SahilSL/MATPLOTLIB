```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]
plt.stem(x,y, linefmt="--", markerfmt='r*')
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]
plt.stem(x,y, linefmt="--", markerfmt='ro')
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]
plt.stem(x,y, linefmt="--", markerfmt='ro', bottom=1)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]
plt.stem(x,y, linefmt="--", markerfmt='ro', bottom=4, label='python')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]
plt.stem(x,y, linefmt="--", markerfmt='ro', bottom=4, label='python', orientation="horizontal")
plt.legend()
plt.show()
```
