```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[1,2,3,4,5]
plt.plot(x,y)
plt.fill_between(x=[2,4],y1=2,y2=4, color='r')
plt.title("Fill Between")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
