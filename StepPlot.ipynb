```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[1,2,3,4,5,6,]
plt.step(x,y)
plt.grid()
plt.title('Step Plot')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[1,2,3,4,5,6,]
plt.step(x,y, marker="o", ms=10, mfc="r", label="Step")
plt.grid()
plt.title('Step Plot')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend()
plt.show()
```
