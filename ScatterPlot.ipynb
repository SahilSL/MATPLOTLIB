```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
plt.scatter(day,no)
plt.title("Scatter Plot", fontsize=20)
plt.xlabel("Days",fontsize=15)
plt.ylabel("Numbers",fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
plt.scatter(day,no, color='r')
plt.title("Scatter Plot", fontsize=20)
plt.xlabel("Days",fontsize=15)
plt.ylabel("Numbers",fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
sizes=[100,200,400,500,200,100,700]
plt.scatter(day,no, s=sizes)
plt.title(",Scatter Plot", fontsize=20),
plt.xlabel("Days",fontsize=15)
plt.ylabel("Numbers",fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
sizes=[100,200,400,500,200,100,700]
plt.scatter(day,no, s=sizes, alpha=0.5)
plt.title(",Scatter Plot", fontsize=20),
plt.xlabel("Days",fontsize=15)
plt.ylabel("Numbers",fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
sizes=[100,200,400,500,200,100,700]
plt.scatter(day,no, s=sizes, alpha=0.5, marker='*')
plt.title(",Scatter Plot", fontsize=20),
plt.xlabel("Days",fontsize=15)
plt.ylabel("Numbers",fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
sizes=[100,200,400,500,200,100,700]
plt.scatter(day,no, s=sizes, alpha=1, marker='+')
plt.title(",Scatter Plot", fontsize=20),
plt.xlabel("Days",fontsize=15)
plt.ylabel("Numbers",fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
sizes=[100,200,400,500,200,100,700]
plt.scatter(day,no, s=sizes, alpha=1, marker='*', edgecolor="black")
plt.title(",Scatter Plot", fontsize=20),
plt.xlabel("Days",fontsize=15)
plt.ylabel("Numbers",fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
sizes=[100,200,400,500,200,100,700]
plt.scatter(day,no, s=sizes, alpha=1, marker='*', edgecolor="black",linewidth=2)
plt.title(",Scatter Plot", fontsize=20),
plt.xlabel("Days",fontsize=15)
plt.ylabel("Numbers",fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
colors=[10,49,30,29,56,20,30]
sizes=[100,200,400,500,200,100,700]
plt.scatter(day,no, s=sizes, c=colors,cmap="viridis")
plt.colorbar()
plt.title("Scatter Plot", fontsize=20)
plt.xlabel("Days",fontsize=15)
plt.ylabel("Numbers",fontsize=15)
plt.show()
```
