```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
area1=[2,3,2,5,4]
area2=[2,3,4,5,6]
area3=[1,3,2,4,2]
l=['A1','A2','A3']
c=['r','g','y']
plt.stackplot(x, area1, area2, area3, labels=l, colors=c)
plt.title('Area Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(loc=2)
plt.show()
```
