```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
vals=[1400,600,300,410,250]
exp_lab=["Home Rent","Food","Phone","Car","Other"]
plt.pie(vals, labels=exp_lab)
plt.show()
```

```python
plt.pie(vals, labels=exp_lab, radius = 2)
plt.show()
```

```python
plt.pie(vals, labels=exp_lab, radius = 2, autopct='%1.1f%%', shadow=True)
plt.show()
```

```python
plt.pie(vals, labels=exp_lab, radius = 2, autopct='%1.1f%%', shadow=True, explode=[0,0.2,0,0.1,0])
plt.show()
```
